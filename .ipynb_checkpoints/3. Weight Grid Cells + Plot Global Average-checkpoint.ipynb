## **Goals**

In this tutorial, we will be going over how to create a global average temperature plot, where the average has been weighted according to the area of each grid cell. If you have already worked through tutorial 1. "Read in Climate Data + Plotting a Time Series" in this repo, you will be familiar with the first few sections - the main difference is that we will be focusing on applying _area weights_ to make the average more representative of the true average temperature over a given region.

_Why weight the grid cells?_

- The Earth is a sphere, so if grid size is specified by angle, that means that grid cells near the poles are physically smaller than those at the equator.
- This happens because the size of the latitudinal circles shrinks as you move toward the poles, based on the cosine of the latitude.
- Weighting ensures these smaller cells are properly represented in the global average.
- When plotting large areas like the global temperature these weights are necessary in order for certain areas not to be artifically over- or undercounted.

<a id='weight'></a>
## **Weight Grid Cells By Area**

As previously stated, when using data over a large area over the globe you need to weight the grid cells in order to ensure accuracy. We will weight the grid cells by creating a weighting function, which we will call `weights`. This function (defined using the Python `def` command) takes as input an xarray DataArray, which is required to have a latitude dimension called `lat`. (You can change this in your function if the file you're using has a different name for the latitude coordinate!)

The weights we need are the areas of each grid cell in the input dataset. In this case, since the data is on a _regular_ grid (equal spacings of latitude and longitude), the cosine of the grid cell latitude works equivalently to the actual spatial area as a weight. The steps carried out by the function are:
- use the Numpy `deg2rad` function to convert the input lat/lon information from degrees to radians, then compute the cosine with Numpy's `cos` function. 
- normalize weights by the mean value over all latitudes
- return the real value of the weighted grid cells

```python
# Define function to generate area weights
def weights(dat):
    # Calculate weighting factor = cosine of latitude
    coslat = np.cos(np.deg2rad(dat.lat))
    weight_factor = coslat / coslat.mean(dim='lat')
    
    # Weight all points by the weighting factor
    computed_weight = dat * weight_factor
    
    # Return the set of weights: this has dimension equal to that of the input data
    return computed_weight
```

After the function is defined, we then apply the newly created function `weights` to the temperature data:

```python
weighted_temp = weights(temp_data)
weighted_temp
```

<a id='global'></a>
## **Calculate the Global Annual Average Temperature**

Next, we'll use the weighted data to calculate the global average, and from that compute the annual average temperature time series. Here all we need to do is average our weighted temperature data over the lat and lon dimensions:

```python
weighted_global_temp = weighted_temp.mean(dim=['lat', 'lon'])
```

We next calculate the annual mean:
-  `.groupby('time.year')` Groups the data by each year.
-  `.mean()` Computes the mean temperature for each year.
- The `yearly_weighted_global_temp` variable now holds the average global temperature for each year, which can be used to observe long-term trends.

```python
yearly_weighted_global_temp = weighted_global_temp.groupby('time.year').mean()
yearly_weighted_global_temp
```

<a id='plot'></a>
## **Plot and Compare Weighted vs Unweighted Global Temperature**

Let's look at the results! If you need a refresher on plotting, please see tutorial 1.

```python
plt.figure()
plt.plot(yearly_weighted_global_temp.year, yearly_weighted_global_temp, label='Weighted Global Mean Temperature (°C)', color='blue')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.title('Global Mean Surface Air Temperature: CanESM5')
plt.legend()
plt.grid()
plt.show()
```

Let's now compare our plot to an unweighted plot and see how much our values vary - to do this, we repeat the spatial and annual averages the same way as above, except skipping the weights function.

```python
unweighted_global_temp = temp_data.mean(dim=['lat', 'lon']) 
unweighted_global_temp = unweighted_global_temp
yearly_unweighted_global_temp = unweighted_global_temp.groupby('time.year').mean() 
```

```python
plt.figure()
plt.plot(yearly_unweighted_global_temp.year, yearly_unweighted_global_temp, label='Unweighted Global Mean Temperature (°C)', color='red')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.title('Global Mean Surface Air Temperature')
plt.legend()
plt.grid()
plt.show()
```

#### We can see here how the temperature values are much lower! This is due to the overcounting of high-latitude data points mentioned at the start of the tutorial.

Now let's add both datasets to a single plot to better visualize the difference due to the weights:

```python
plt.figure()
plt.plot(yearly_weighted_global_temp.year, yearly_weighted_global_temp, label='Weighted Global Mean Temperature (°C)', color='blue')
plt.plot(yearly_unweighted_global_temp.year, yearly_unweighted_global_temp, label='Unweighted Global Mean Temperature (°C)', color='red')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.title('Global Mean Surface Air Temperature: CanESM5')
plt.legend()
plt.grid()
plt.show()
```

The creation of weighted averages is something you can now keep in mind when calculating your own regionally averaged time series!
